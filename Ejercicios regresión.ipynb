```python
import numpy as ny
```

```python
gamma = ny.array([50,70,90,110,130])
thau = ny.array([6.01,7.48,8.59,9.19,10.21])
```

#### A partir de los datos se realizará un ajuste de ecuación de potencia de la forma:
