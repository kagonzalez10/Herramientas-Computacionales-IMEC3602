```python
gamma = ny.array([50,70,90,110,130])
thau = ny.array([6.01,7.48,8.59,9.19,10.21])
```

#### A partir de los datos se realizar치 un ajuste de ecuaci칩n de potencia de la forma:
$$\tau = \mu\gamma^n$$

```python

plt.plot(thau,gamma,'*k',label='Datos')
plt.xlabel('$\dot\gamma$')
plt.ylabel('$\\tau$')
plt.legend()
plt.show()
```

#### Se realizar치 una regresi칩n lineal para los datos anteriormente mostrados:

```python
(a,b)= polyfit(gamma,thau,1)
print('Regresi칩n lineal con par치metros: a=%.2f b=%.2f' % (a,b))
```

```python
t = np.linspace(gamma[0],gamma[-1])
modelo = polyval([a,b],t)
plt.plot(gamma,thau,'*b',label = 'Datos')
plt.plot(t,modelo,'-k', label = 'a=%.2f b=%.2f' % (a,b))
plt.legend()
```
