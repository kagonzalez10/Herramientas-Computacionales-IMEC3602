```python
%pylab inline
from scipy.integrate import odeint
import os
import pandas as pd
from scipy import ndimage
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
#### DATOS DE KART

# Datos Llantas tractivas 
rll=0.139 # Radio geométrico de las llantas en traseras (m)
refc=0.98*rll # Radio efectivo llantas traseras (m)
rin=0.0631 #radio interno del neumático (m)

# Datos Transmisión
dp=11.0 # Dientes del piñon
dpl=39.0 # Dientes del plato
ng=dpl/dp # Relación de Transmisión
etaf=0.95 # Eficiencia de transmisión

# Datos eje llantas
reje=0.02 # Radio eje de las llantas (m)
meje=3.832 # Masa del eje (kg)

# Datos Kart general
masa=152.0 # masa del Kart (kg)
g=9.81 # Gravedad (m/s)
w=masa*g # peso del Kart (N)

# Datos Fuerzas de la fórmula de aceleración
rho=1.1955 # densidad del aire (kg/m^3)
dll=0.975 # Distancia entre ejes de llantas delanteras (m)
alj=0.88 # Altura con el piloto sentado en el kart (estatura 1.68 m) (m)
af=(dll*alj)/2 # Area Frontal Kart (m^2)
cd=0.89 # Coeficiente de Arrastre Kart
vw=0.0 # velocidad del viento (m/s)
theta=0.0 # Angulo carretera

# Inercias
ii=5.2E-3 # Inercia motor de combustión (kg m^2) 

mllt=1.43 # Masa llantas traseras (kg)
mlld=1.20 # Masa llantas delanteras (kg)
rlld=0.129 # Radio llantas delanteras (m)
rind=0.0631 # Radio rin llantas delanteras (m)
mbt=0.7*mllt # El porcentaje de la masa total que pesa la parte "banda" de la llanta trasera
mdt=0.3*mllt # El porcentaje de la masa total que pesa la parte "disco" de la llanta trasera
mbd=0.7*mlld # El porcentaje de la masa total que pesa la parte "banda" de la llanta delantera
mdd=0.3*mlld # El porcentaje de la masa total que pesa la parte "disco" de la llanta delantera
illd=(mbd*rlld)+(mdd*(rlld**2-rind**2))/2 # Inercia 1 llanta delantera (kg-m^2)
illt=(mbt*rll)+(mdt*(rll**2-rin**2))/2 # Inercia 1 llanta trasera (kg-m^2)

#paso=7.62 #(mm)
#rpim= paso*(0.6+(1/tan(180/dp))) # Radio Piñon (mm)
#rplam=paso*(0.6+(1/tan(180/dpl))) # Radio Plato (mm)
#rpi= rpim/1000 #Radio Piñon (m)
#rpla= rplam/1000 #Radio Plato (m)
rpi=0.01525#(m)
rpla=0.04096 #(m)
rhoac=7850.0 # Densidad del Acero (kg/m^3)
bpi=0.01 # Ancho piñon (m)
bpla=0.02 # Ancho plato (m)
vpi= bpi*np.pi*rpi**2 # Volumen del piñon (m^3)
vpla=bpla*np.pi*rpla**2 # Volumen del plato (m^3)
mpi=rhoac*vpi # Masa del piñon (kg)
mpla=rhoac*vpla # Masa del plato (kg)
ig=((mpi*(rpi**2))/2.0)+((mpla*(rpla**2))/2.0) # Inercia de la transmisión (kg-m^2)

ieje=(meje*reje**2)/2.0 # Inercia del eje de las llantas (kg-m^2)

# Fuerzas Aceleración
# Resistencia a la rodadura
a=0.2231 # Distancia del centro de gravedad al eje de la llanta trasera (m)
l=1.0063 # Distancia de eje de las llantas traseras y delanteras (m)
h=0.07478 # Altura del centro de gravedad del kart con respecto al piso (m)

def accel(vx, ddt):
    return ndimage.gaussian_filter1d(vx, sigma=20, order=1, mode='wrap') / ddt

def frodadura(vx,ax):
    wr=w*((a/l)+(ax/g*(h/l))) # fuerza vertical sobre las llantas tractivas
    fr=0.01*(1+(vx/100.0)) # Factor de rodadura la velocidad del kart debe estar en (km/h)
    # fr=C*((mllt*g)/(rllt*2))*sqrt(ht/bllt)
    # ht=0.0739 # Altura del neumático (m)
    # bllt=0.209 # Ancho del neumático (m)
    # C= # Constante del material de la llanta
    return wr*fr*cos(theta) # (N)

# Fuerza del ángulo de la carretera
def finclinacion(theta):
    return w*sin(theta) # (N)

# Fuerza de arrastre
def farrastre(vx):
    return (1.0/2.0)*rho*af*cd*(vx+vw)**2 # (N)

# Fuerza de tracción, Torque del Motor
def ftraccion(ax,Fg,Fw,Fr):
    # Ft=((ti*ng*etaf)/refc)-(((ii-ig)*ng^2)+ieje+illd*2+illt*2)*(ax/(refc^2)) # (N)
    return masa*ax+Fg+Fw+Fr

def torquemotor(ax,Ft):
    return (Ft+(((ii-ig)/(ng**2))+ieje+illd*2+illt*2)*(ax/(refc**2)))*(refc/(ng*etaf)) # Torque del motor (Nm)

# Aceleración, Fuerza de tracción
# mr=(((ii-ig)*ng^2)+ieje+illd*2+illt*2)/(refc^2) 
# ax=(((ti*ng*etaf)/refc)-Fg-Fw-Fr)/(m+mr) # (m/s)

# Velocidad angular del motor
def vangularmotor(vx):
    return (vx*ng)/refc # (rad/s)
```

```python
### Datos de motor eléctrico
def DataFrameE_fileNumber(numE):
    data_file = './data_electrico/VBS__0'+numE+'.VBO'
    with open(data_file, 'r',encoding='cp1252') as vboxdata:
        sectione = []
        timee = []
        late = []
        longe = []
        vel = []
        heighte = []
        vertve = []
        for line in vboxdata:
            if line[0] == '[':
                sectione = str.replace(line, "\n","")
                continue
            if sectione == '[column names]':
                col_names = str.replace(line, "  \n","").split('  ')
                section = []
            if sectione == '[data]':
                linesplit = str.replace(line, " \n","").split(' ')
                timee.append(float(linesplit[1][0:2])*3600+float(linesplit[1][2:4])*60+float(linesplit[1][4:]))
                late.append(float(linesplit[2]))
                longe.append(float(linesplit[3]))
                vel.append(float(linesplit[4]))
                heighte.append(float(linesplit[6]))
                vertve.append(float(linesplit[7]))

    timee = np.asarray(timee)
    late = np.asarray(late)
    longe = np.asarray(longe)
    vele = np.asarray(vel)/3.6
    heighte = np.asarray(heighte)
    vertve = np.asarray(vertve)
    time_basee = timee[0]
    timee = timee - time_basee
    late = late/60.0
    longe = -longe/60.0
    dte = timee[1]-timee[0]
###DATA FRAME Motor Eléctrico
    datose = pd.DataFrame(columns=['Tiempo','Lat','Long','Velocidad','Height','Vertv'])
    datose['Tiempo']=timee
    datose['Lat']=late
    datose['Long']=longe
    datose['Velocidad']=vele
    datose['Height']=heighte
    datose['Vertv']=vertve
    datose['Aceleracion']= accel(datose.loc[:,'Velocidad'],dte)
    return datose
#### Datos motor de combustión
def DataFrameC_fileNumber(numC):
    data_file = './data_combustion/VBOX00'+numC+'.VBO'
    with open(data_file, 'r',encoding='cp1252') as vboxdata:
        section = []
        time = []
        lat = []
        long = []
        vel = []
        height = []
        vertv = []
        for line in vboxdata:
            if line[0] == '[':
                section = str.replace(line, "\n","")
                continue
            if section == '[column names]':
                col_names = str.replace(line, "  \n","").split('  ')
                section = []
            if section == '[data]':
                linesplit = str.replace(line, " \n","").split(' ')
                time.append(float(linesplit[1][0:2])*3600+float(linesplit[1][2:4])*60+float(linesplit[1][4:]))
                lat.append(float(linesplit[2]))
                long.append(float(linesplit[3]))
                vel.append(float(linesplit[4]))
                height.append(float(linesplit[6]))
                vertv.append(float(linesplit[7]))

    time = np.asarray(time)
    lat = np.asarray(lat)
    long = np.asarray(long)
    velc = np.asarray(vel)/3.6
    height = np.asarray(height)
    vertv = np.asarray(vertv)
    time_base = time[0]
    time = time - time_base
    lat = lat/60.0
    long = -long/60.0
    dt = time[1]-time[0]
    ###DATA FRAME Motor de Combustión
    datosc = pd.DataFrame(columns=['Tiempo','Lat','Long','Velocidad','Height','Vertv'])
    datosc['Tiempo']=time
    datosc['Lat']=lat
    datosc['Long']=long
    datosc['Velocidad']=velc
    datosc['Height']=height
    datosc['Vertv']=vertv
    datosc['Aceleracion']= accel(datosc.loc[:,'Velocidad'],dt)
    return datosc
```

```python
def plotting_vel (datos):
    plt.plot(datos.loc[:,'Tiempo'],datos.loc[:,'Velocidad'],'k')
    xlabel('Tiempo(s)')
    ylabel('Velocidad(m/s)')
    title('Velocidad Vs. Tiempo')
    pylab.legend(loc='upper left')
```

```python
def plotting_vel2 (datosc,datose):
    plt.figure(figsize=(10,5))
    plt.plot(datosc.loc[:,'Tiempo'],datosc.loc[:,'Velocidad'],'k',label= 'Combustión')
    plt.plot(datose.loc[:,'Tiempo'],datose.loc[:,'Velocidad'],'g',label= 'Eléctrico')
    xlabel('Tiempo($s$)')
    ylabel('Velocidad($m/s$)')
    title('Velocidad Vs. Tiempo')
    pylab.legend(loc='best')
```

```python
def plotting_velFrag (datosc,datose,inicioE,finE,inicioC,finC):
    plt.figure(figsize=(10,5))
    plt.plot(datosc.loc[:,'Tiempo'][inicioC:finC]-datosc.loc[:,'Tiempo'][inicioC],datosc.loc[:,'Velocidad'][inicioC:finC],'k',label= 'Combustión')
    plt.plot(datose.loc[:,'Tiempo'][inicioE:finE]-datose.loc[:,'Tiempo'][inicioE],datose.loc[:,'Velocidad'][inicioE:finE],'g',label= 'Eléctrico')
    xlabel('Tiempo($s$)')
    ylabel('Velocidad($m/s$)')
    title('Velocidad Vs. Tiempo')
    pylab.legend(loc='best')
```

```python
def plotting_AcelFrag (datosc,datose,inicioE,finE,inicioC,finC):
    plt.figure(figsize=(10,5))
    plt.plot(datosc.loc[:,'Tiempo'][inicioC:finC]-datosc.loc[:,'Tiempo'][inicioC],datosc.loc[:,'Aceleracion'][inicioC:finC],'k',label= 'Combustión')
    plt.plot(datose.loc[:,'Tiempo'][inicioE:finE]-datose.loc[:,'Tiempo'][inicioE],datose.loc[:,'Aceleracion'][inicioE:finE],'g',label= 'Eléctrico')
    xlabel('Tiempo($s$)')
    ylabel('Aceleración($m/s^{2}$)')
    title('Aceleración Vs. Tiempo')
    pylab.legend(loc='best')
```

```python
def plotting_acel (datosc,datose):
    plt.figure(figsize=(10,5))
    plt.plot(datosc.loc[:,'Tiempo'],datosc.loc[:,'Aceleracion'],'k',label= 'Combustión')
    plt.plot(datose.loc[:,'Tiempo'],datose.loc[:,'Aceleracion'],'g',label='Eléctrico')
    xlabel('Tiempo($s$)')
    ylabel('Aceleracion($m/s^{2}$)')
    title('Aceleracion Vs. Tiempo')
    pylab.legend(loc='best')
```
